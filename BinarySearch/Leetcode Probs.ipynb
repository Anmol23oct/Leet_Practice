```python
class Solution(object):
    def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        a=nums[0]
        last=len(nums)-1
        b=nums[last]
        if a<=b:
            return a
        else:
            j=last
            while b>=nums[j]:
                if b==nums[j]:
                    j=j-1
                else:
                    b=nums[j]
                    j=j-1
            return b
                
        


s=Solution()
n=input()
nums=list(map(int,n.split()))
print(s.findMin(nums))
```

# Contains Duplicate

```python
class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        nums.sort()
        for i in range(len(nums)):
            if i+1!=len(nums):
                if nums[i]==nums[i+1]:
                    return True
        
        return False
                
        


s=Solution()
n=input()
nums=list(map(int,n.split()))
print(s.containsDuplicate(nums))
```

# Product of Array Except Self

```python
class Solution(object):
    def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        left=[]
        prodl=1
        for i in range(len(nums)):
            prodl=prodl*nums[i]
            left.append(prodl)
        
        right=[]
        prodr=1
        for i in range(len(nums)-1,-1,-1):
            prodr=prodr*nums[i]
            right.append(prodr)
        
        #print("left",left)
        rightf=right[::-1]
        #print("right",rightf)
        final=[]
        for i in range(len(nums)):
            if i==0:
                if i+1<len(nums):
                    final.append(rightf[i+1])
            elif i==len(nums)-1:
                if i-1>-1:
                    final.append(left[i-1])
            else:
                final.append(left[i-1]*rightf[i+1])
        
        return final







s=Solution()
n=input()
nums=list(map(int,n.split()))
print(s.productExceptSelf(nums))
```

# Min Cost Path Problem

```python
import sys
from sys import stdin
MAX_VALUE = 2147483647

def minCostPath(mat,i,j, n, m,dp) :
	#Your code goes here
    if (i==n-1) and (j==m-1):
        return mat[i][j]
    
    if i>=n or j>=m:
        return sys.maxsize
    
    if dp[i][j+1]==sys.maxsize:
        
        right=minCostPath(mat,i,j+1,n,m,dp)
        dp[i][j+1]=right
    
    else:
        right=dp[i][j+1]
    
    if dp[i+1][j+1]==sys.maxsize:
        
        slant=minCostPath(mat,i+1,j+1,n,m,dp)
        dp[i+1][j+1]=slant
    
    else:
        slant=dp[i+1][j+1]
    
    if dp[i+1][j]==sys.maxsize:
        
        down=minCostPath(mat,i+1,j,n,m,dp)
        dp[i+1][j]=down
    
    else:
        down= dp[i+1][j]
    
    
    fCost=mat[i][j]+min(right,slant,down)
    
    return fCost



























#taking input using fast I/O method
def take2DInput() :
    li = input().split(" ")
    n = int(li[0])
    m = int(li[1])
    
    if n == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(n)]
    return mat, n, m


#main
mat, n, m = take2DInput()
dp=[[sys.maxsize for j in range(m+1)] for i in range(n+1)]
print(minCostPath(mat,0,0, n, m,dp))
```
